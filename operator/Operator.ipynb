Question: Calculate the floor division of -17 by 5. Explain why the result is -4 and not -3.

```python
result = -17 // 5
print(result)
```

```python
import math

result = math.floor(-17 / 5)
print(result)
```

```python
import math

division = -17 / 5
result = int(math.floor(division))
print(result)
```

```python
a = -17
b = 5

result = (a - (a % b)) // b # a - (a % b)  -17 - 3 = -20   ,  -20 // 5 = -4
print(result)
```

```python
division = -17 / 5 # -3.4

if division < int(division):   # -3    ,  -3 > -3.4
    result = int(division) - 1  # -4
else:
    result = int(division)
print(result)
```

Question: Calculate the remainder of -17 divided by 5. In Python, the remainder takes the sign of the divisor.

```python
result = -17 % 5
print(result)
```

```python
a = -17
b = 5

result = a - (a // b) * b
print(result)
```

```python
quotient, remainder = divmod(-17 , 5) # a // b , a % b  #  red = a - (b * quotient) = -17 - (5 * -4) = -17 - (-20) = 3
print(remainder)
```

```python
a = -17
b = 5

remainder = b - (abs(a) % b) # b = 5 , a = -17 , abs(a) = abs(-17) = 17  , 17 / 5 = 3 , 17 % 5 = 2 , 5 - 2 = 3
print(remainder)
```

```python
a = -17
b = 5

while a < 0:  # 3 < 0
    a += b
result = a % b # 3 % 5 = 3
print(result)


# a = -17 , a += b = -17 + 5 = -12 , -12 + 5 = -7 , -7 + 5 = -2 , -2 + 5 = 3
```

Question: Initialize x = 10. Update x by multiplying it by 3 + 2 using the compound assignment operator *=. Prove that addition happens before multiplication.

```python
x = 10
x *= 3 + 2 # x 10 * 5 =50
print(x)
```

```python
x = 10
temp = 3 + 2
x *= temp
print(x)
```

```python
x = 10
x = x * (3 + 2) # x = 10 * 5 = 50
print(x)
```

```python
x = 10
x = (x * 3) + 2
print(x) # 50 != 32
```

```python
# eval

x = 10
expression = "3 + 2" # text
value = eval(expression) # eval ---> text ---> (number) = 5
x *= value # x = 5 * 10 = 50
print(x)
```
